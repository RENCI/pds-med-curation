```python
# Import dependencies

import requests
import json
import pandas as pd
from pprint import pprint
```

```python
# Open MedicationRequest.keep.code-class text file

file = open("MedicationRequest.keep.code-display-system", "r")
```

```python
rxcodes = []
```

```python
# Read the file

lines = file.readlines()
```

```python
# Extract drug rxcui code from each row

for line in lines:
    rxcodes.append(line.split('\t'))
```

```python
# Remove drug class from the list
# rxcodes has only drug codes

for r in rxcodes:
    r.pop(1)
```

```python
# Convert rxcodes list to dataframe 

df = pd.DataFrame(rxcodes)
```

```python
# Retain unique rxcui codes

query_code = df[0].unique()
```

```python
query_code.size
```

```python
# Base query URL

base_url = "https://rxnav.nlm.nih.gov/REST/rxcui/"
```

```python
i=0
ilist = {}
for code in query_code:
    # Complete query URL
    query_url = base_url+code+"/related.json?tty=IN"
    response = requests.get(query_url)
    data = response.json()
    try:
        rel_props = data['relatedGroup']['conceptGroup'][0]['conceptProperties']
        umlscuis = [ e['umlscui'] for e in rel_props ] 
        ilist.update({code:umlscuis})
    except:
        pass
```

```python
ilist
```

```python
with open('rxnorm_map.json', 'w') as json_file:
    json.dump(ilist, json_file)
```
